## Preprocess and pickle the test data

```python
for p in glob.glob('test_adorned_xml/*.xml'):
    
    file_name = p.split('/')[-1].split('.')[0]
    
    file_data = preprocess_one_file(p, SHINGLE_LENGTH)

    f = open('test_pickles/' + file_name + '.pickle', 'wb')
    pickle.dump(file_data, f)
    f.close()
    
print('Done!')
```

### Match a partial *Shepheardes Calender* to a partial *England's Helicon*

These results were desked checked, and are reasonable.

```python
import textwrap

SHINGLE_LENGTH = 3
MAX_GAP_ALLOWED = 5
MIN_MATCH_LENGTH = 6
     
file_a = load_pickle_file('test_pickles/A12782_test.pickle')
file_b = load_pickle_file('test_pickles/A16274_test.pickle')

results = match_two_files(file_a, 
                            file_b,
                            MAX_GAP_ALLOWED, MIN_MATCH_LENGTH)
    
for fn, f in enumerate(results):
    print(fn, '----------------------------------------------------------------------------')
    print(textwrap.fill(f[0], 65))
    print()
    print(textwrap.fill(f[1], 65))
```
