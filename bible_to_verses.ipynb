## Turn non-TCP XML into "Bible Pickles"

I originally thought that it would be best to use non-EEBO-TCP Bibles as the source for verse-by-verse matching.  EEBO-TCP have a lot of gaps in transcription (10,000 in its Geneva?  15,000?), which I imagined would cause me to miss matches.  Hence this process.

However, as it turns out, the EEBO-TCP actually seems to match better that the files processed here.  Mostly, I suspect, because the matching process is built to work around the problems of gaps.

### SHINGLE_LENGTH . . . 

 . . . is the key variable here.  In effect, it's the number of non-stopword, non-space, non punctuation lemma which constitutes the minimun span for matching. For more, please see matching_explained.ipynb.

### Process the KJV from the Oxford Text Archive

These processes are not completely generalizable because, although these documents are all in TEI, they all mark up books, chapters, and verses differently.  Note, however, that the low-level functions, like shingling, are generalized.

This bit of code

    results.append({'reference': reference, 
                    'tokens': tokens, 
                    'lemmas': lemmas,
                    'non_space_lemmas': non_space_lemmas, 
                    'offsets': offsets,
                    'shingles': shingles})
                  
is a fairly good, high-level representation of the contents of one verse in a "bible pickle."

```python
tree = etree.parse('adorned_bible/KJV_OTA.xml')

books = tree.xpath('//tei:div[@type="book"]', namespaces={'tei': 'http://www.tei-c.org/ns/1.0'})

results = []

for b in books:
    
    book_id = b.get('{http://www.w3.org/XML/1998/namespace}id').upper()
    head = b.xpath('descendant::tei:head', namespaces={'tei': 'http://www.tei-c.org/ns/1.0'})[0]
    title = re.sub('\s+', ' ', ''.join(head.itertext()).strip()).split(':')[-1].strip().replace(' ', '_')
    
    print(title)
    
    for p in b.xpath('descendant::tei:p', namespaces={'tei': 'http://www.tei-c.org/ns/1.0'}):
        
        reference = title + '.' + '.'.join(p.get('n').split('.')[1:])
    
        tokens, lemmas = get_tokens_for_iterator(p)
        non_space_lemmas, offsets = get_non_space_lemmas_and_offsets(lemmas)
        shingles = shingle_tokens(non_space_lemmas, SHINGLE_LENGTH)
        
        results.append({'reference': reference, 
                        'tokens': tokens, 
                        'lemmas': lemmas,
                        'non_space_lemmas': non_space_lemmas, 
                        'offsets': offsets,
                        'shingles': shingles})

f = open('bible_pickles/KJV_OTA.pickle', 'wb')
pickle.dump(results, f)
f.close()
```

### Process Geneva

```python
import codecs, re

for p in glob.glob('adorned_bible/geneva/*.xml'):
    
    file_name = p.split('/')[-1].split('.')[0]
    
    data = codecs.open(p, 'r', encoding='utf-8').read()
    data = re.sub(' xml\:id\="[0-9A-Za-z\-\_\.]+"', '', data)
    
    tree = etree.XML(data)
    
    title = tree.xpath('//tei:book', namespaces={'tei': 'http://www.tei-c.org/ns/1.0'})[0].get('name')
    title = title.replace(' ', '_')
    
    print(title)

    results = []
    
    for chapter in tree.xpath('//tei:chapter', namespaces={'tei': 'http://www.tei-c.org/ns/1.0'}):
        
        chapter_n = chapter.get('number')
        
        for verse in chapter.xpath('descendant::tei:verse', namespaces={'tei': 'http://www.tei-c.org/ns/1.0'}):
        
            verse_n = verse.get('number')
            
            reference = title + '.' + chapter_n + '.' + verse_n
    
            tokens, lemmas = get_tokens_for_iterator(verse)
            non_space_lemmas, offsets = get_non_space_lemmas_and_offsets(lemmas)
            shingles = shingle_tokens(non_space_lemmas, SHINGLE_LENGTH)

            results.append({'reference': reference, 
                            'tokens': tokens, 
                            'lemmas': lemmas,
                            'non_space_lemmas': non_space_lemmas, 
                            'offsets': offsets,
                            'shingles': shingles})

    f = open('bible_pickles/geneva/' + file_name + '.pickle', 'wb')
    pickle.dump(results, f)
    f.close()
```

### Process the KJV Apocrypha

```python
import codecs, re

for p in glob.glob('adorned_bible/KJV_apocrypha/*.xml'):
    
    file_name = p.split('/')[-1].split('.')[0]
    
    print(file_name, p)
    
    data = codecs.open(p, 'r', encoding='utf-8').read()
    data = re.sub(' xml\:id\="[0-9A-Za-z\-\_\.]+"', '', data)
    
    tree = etree.XML(data)
    
    heading = None
    chapter = None
    verses = []
    
    for token in tree.xpath('//tei:w|//tei:pc|//tei:c', namespaces={'tei': 'http://www.tei-c.org/ns/1.0'}):
        
        p = token.getparent()
        
        if p.tag == '{http://www.tei-c.org/ns/1.0}span':
            
            if p.get('class') == 'heading':
                heading = token.text
            elif p.get('class') == 'title':
                pass
            elif p.get('class') == 'chapter':
                chapter = token.text.strip()
            elif p.get('class') == 'verse':
                verses.append([chapter, token.text, []])
            else:
                verses[-1][2].append(token)
         
    results = []
        
    for v in verses:
        
        reference = (heading + '.' + v[0] + '.' +  v[1])
        
        tokens = []
        lemmas = []
        
        for token in v[2]:
            
            selected_token, selected_lemma = select_token_and_lemma(token, 'spe')
            
            if selected_token != None and selected_lemma != None:
                tokens.append(selected_token)
                lemmas.append(selected_lemma)
                    
        non_space_lemmas, offsets = get_non_space_lemmas_and_offsets(lemmas)
        shingles = shingle_tokens(non_space_lemmas, SHINGLE_LENGTH)

        results.append({'reference': reference, 
                        'tokens': tokens, 
                        'lemmas': lemmas,
                        'non_space_lemmas': non_space_lemmas, 
                        'offsets': offsets,
                        'shingles': shingles})

    f = open('bible_pickles/KJV_apocrypha/' + file_name + '.pickle', 'wb')
    pickle.dump(results, f)
    f.close()
```
